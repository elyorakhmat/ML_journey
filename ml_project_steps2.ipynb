````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
````

````python
df = pd.read_csv('https://github.com/ageron/handson-ml2/blob/master/datasets/housing/housing.csv?raw=true')
df.head()
````

````python
df.info()
````

````python
df.isna().sum()
````

# We can do the following actions to handle NaN values:
* Removing all rows with NaN values.
* Droping the entire column with NaN values.
* Repacing NaN values with mean values

````python
# 1
df.dropna(subset= ['total_bedrooms'])
````

````python
# 2
df.drop('total_bedrooms', axis=1)
````

````python
# 3
mean = df['total_bedrooms'].mean()
df['total_bedrooms'].fillna(mean)
````

# Dealing with NaN values by using sklearn methods: transformers

````python
from sklearn.impute import SimpleImputer
````

````python
imputer = SimpleImputer(strategy= 'mean')
````

````python
df_num = df.drop('ocean_proximity', axis= 1)
df_num
````

# Calculating mean vaues of each column

````python
imputer.fit(df_num)
````

# printing the statistics (mean)

````python
imputer.statistics_
````

# replacing NaN values with the mean values

````python
x= imputer.transform(df_num)
x
````

# Doing both tasks at once by ```.fit_transform()``` method

````python
x= imputer.fit_transform(df_num)
x
````

# We can return the array values above to data frame


````python
df_tr = pd.DataFrame(x, columns= df_num.columns, index= df_num.index)
df_tr.head()
````

````python
df_tr['ocean_proximity'] = df['ocean_proximity']
df_tr.head()
````

````python
df_tr.to_csv('housing.csv', index= False)
````
