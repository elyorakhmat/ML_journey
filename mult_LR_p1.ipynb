# Multiple Linear Regression practice.
We will develop a MLR model that can predict house prices by their features.

```python
# libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/housing_LR.csv', index_col=0)
df.head()
```

```python
# separating the wanted dataset
housing = df[df['district'] == 'Чиланзарский']
```

```python
# correlations
housing.corr(numeric_only= True)
```

# The scinario is similar to that of simple linear regression, but the x prameters (features) are multiple.


```python
# dataset train-test splitting.
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(housing, test_size= 0.1, random_state= 42)
```

```python
# features and labels
# train set
x_train = train_set[['size', 'max_levels']].to_numpy()
y_train = train_set[['price']].to_numpy()

# test set
x_test = test_set[['size', 'max_levels']].to_numpy()
y_test = test_set[['price']].to_numpy()
```

```python
# model training
from sklearn import linear_model
MLR_model = linear_model.LinearRegression()

MLR_model.fit(x_train, y_train)
```

```python
# predicting
y_pred = MLR_model.predict(x_test)
```

```python
# calculating coeffitcients
# coeffitcients
coeffs = MLR_model.coef_[0]

# intercept
b = MLR_model.intercept_

print('Coeffitcients: ', coeffs)
print('Intercept: ', b)
```

```python
# comparing
comp_show = pd.DataFrame({'Predicted': y_pred.flatten(), 'Real': y_test.flatten()})
comp_show.head()
```

```python
# evaluation
from sklearn.metrics import mean_absolute_error, mean_squared_error

# MAE
mae = mean_absolute_error(y_pred, y_test)

# RMSE
rmse = np.sqrt(mean_squared_error(y_pred, y_test))

print('MAE = ', mae)
print('RMSE = ', rmse)
```
