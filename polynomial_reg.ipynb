# A simple linear regression model for polynomial trend.

```python
# libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%202/data/china_gdp.csv")
df.head()
```

```python
# visualization
plt.plot(df['Year'], df['Value'], 'go')
plt.show()
```

```python
# normalization
X = (df[['Year']]/df['Year'].max()).to_numpy()
y = (df[['Value']]/df['Value'].max()).to_numpy()
```

```python
# building a simple linear reg model
from sklearn import linear_model

SLR_model = linear_model.LinearRegression()

SLR_model.fit(X, y)
```

```python
# prediction
y_pred = SLR_model.predict(X)
```

```python
# graphic
plt.plot(X, y, 'go')
plt.plot(X, y_pred, 'r-')
plt.show()
```

```python
# model evaluation
from sklearn.metrics import mean_absolute_error
# MAE
mae = mean_absolute_error(y, y_pred)

print('MAE = ', mae)
```

# A polynomial reg model (Linear reg).

```python
# extracting polynomial features
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree= 5, include_bias= False)
X_poly = poly_features.fit_transform(X)
```

```python
X_poly
```

```python
# developing a PR model
PR_model = linear_model.LinearRegression()

PR_model.fit(X_poly, y)
```

```python
# prediction
y_pred = PR_model.predict(X_poly)
```

```python
# vizualization
plt.plot(X, y, 'go')
plt.plot(X_poly, y_pred, 'r-')
plt.show()
```

```python
# evaluation
# MAE
mae = mean_absolute_error(y_pred, y)

print('MAE = ', mae)
```
