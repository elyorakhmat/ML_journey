# Min-Max Normalization
Putting values between 0 and 1
The formula of min-max normalization:
# xnew = (x - xmin) / (xmax - xmin)

```python
import numpy as np
import pandas as pd
import sklearn
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/elyorbek8/ML_journey/refs/heads/main/housing.csv')
df.head()
```

```python
df.describe()
```

# Calculating manually

```python
min_income = df['median_income'].min()
min_income
```

```python
max_income = df['median_income'].max()
max_income
```

```python
df.loc[134, 'median_income']
```

```python
xnew = (df.loc[134, 'median_income'] - min_income) / (max_income - min_income)
xnew
```

# Calculating by a special sklearn model


```python
df_num = df.drop('ocean_proximity', axis= 1)
df_num.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()

min_max_scaler.fit_transform(df_num)
```

# Standardization
Formula: x_new = (x - x_mean) / std

std = standard deviation

```python
from sklearn.preprocessing import StandardScaler

standard_scaler = StandardScaler()
standard_scaler.fit_transform(df_num)
```
