# Encoding: tranforming characterical values to a numeric form

# Ordinal Encoding

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
````

````python
df = pd.read_csv('https://raw.githubusercontent.com/elyorbek8/ML_journey/refs/heads/main/housing.csv')
df.head(10)
````

````python
ocean_prox = df['ocean_proximity']
ocean_prox.head()
````

````python
ocean_prox.value_counts()
````

````python
ocean_prox.info()
````

````python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()

ocean_prox_encoded = ordinal_encoder.fit_transform(ocean_prox.values.reshape(-1, 1))
ocean_prox_encoded
````

````python
ocean_prox_encoded[[6244, 5634, 19813, 7299, 6648, 16329, 11600, 5673, 6787, 10052]]
````

# One hot encoding: ```sklearn OneHotEncoder```

````python
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder = OneHotEncoder()
````

````python
ocean_prox_1hot = one_hot_encoder.fit_transform(ocean_prox.values.reshape(-1, 1))
ocean_prox_1hot.toarray()
````

# One hot encoding: ```pandas get_dummies()```

````python
ocean_prox_dummies = pd.get_dummies(ocean_prox)
ocean_prox_dummies.head()
````
